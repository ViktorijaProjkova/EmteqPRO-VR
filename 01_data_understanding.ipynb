```python

def extract_participant_id(folder_name):
    # baraj broj vo imeto na folderot
    match = re.search(r"\d+", folder_name)
    # \d e bilo koja cifra
    return int(match.group()) if match else 0

# za site participants lista od DATA_DIR
# site sto se so participant
participant_dirs = [
    p for p in DATA_DIR.iterdir()
    if p.is_dir() and p.name.lower().startswith("participant")
]

# sortiraj gi po rastecki redosled
participant_dirs = sorted(participant_dirs, key=lambda p: extract_participant_id(p.name))

# br na participants
print("Number of participant folders:", len(participant_dirs))

print("First 10:", [p.name for p in participant_dirs[:10]])

print("Last 10:", [p.name for p in participant_dirs[-10:]])
```

```python
#naogjanje mozni vazni koloni 
keywords = [
    "participant", "subject", "user", "id",
    "video", "stimulus", "clip", "movie",
    "valence", "val",
    "arousal", "aro",
    "emotion",
    "time", "timestamp", "start", "end", "duration"
]

def find_matching_columns(df, keywords):
    matches = []
    for col in df.columns:
        col_lower = str(col).lower()
        if any(k in col_lower for k in keywords):
            matches.append(col)
    return matches

print("Possible important columns in DETAILS:")
print(find_matching_columns(details, keywords))

print("\nPossible important columns in QUESTIONNAIRES:")
print(find_matching_columns(questionnaires, keywords))
```

## 5. Inspect participant CSV files

- Open one participant folder to see the sensor and annotation files inside it.
- Preview the first rows of each CSV to understand their schema.
- This reveals the structure of the raw per-participant data that will later be used for feature extraction.

```python

def read_csv_auto(path, nrows=None):
    # ovie separatori
    separators = [",", ";", "\t"]
    encodings = ["utf-8-sig", "utf-8", "latin1"]

    for encoding in encodings:
        for sep in separators:
            try:
                # obid za citanje csv so tekovnite parametri
                df = pd.read_csv(
                    path,
                    sep=sep,
                    encoding=encoding,
                    nrows=nrows,
                    low_memory=False
                )

                if df.shape[1] > 1:
                    return df
            except Exception:
                pass

    # ako ne pomine niedna kombinacija, probaj najblaga verzija
    return pd.read_csv(
        path,
        sep=",",
        encoding="latin1",
        nrows=nrows,
        on_bad_lines="skip",
        low_memory=False
    )
```

```python
#kje gi otvori prvite 5 reda od sekoj csv 
for f in sample_files:
    print("\n====================================")
    print("FILE:", f.name)
    
    df_sample = read_csv_auto(f, nrows=5)
    
    print("Shape:", df_sample.shape)
    print("Columns:")
    print(df_sample.columns.tolist())
    
    display(df_sample.head())
```

```python
sensor_keywords = [
    "time", "timestamp", "video", "stimulus",
    "event", "segment", "label", "start", "end"
]

for f in sample_files:
    df_sample = read_csv_auto(f, nrows=5)
    
    matching_cols = [
        col for col in df_sample.columns
        if any(k in str(col).lower() for k in sensor_keywords)
    ]
    
    print(f.name, "=>", matching_cols)
```
